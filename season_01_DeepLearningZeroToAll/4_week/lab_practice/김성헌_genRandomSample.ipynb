# 머신러닝 학습에 사용할 랜덤 데이터 생성

```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import pandas as pd

# 2차원 random 샘플 데이터 생성
mu_vec1 = np.array([0,0])
cov_mat1 = np.array([[2,0],[0,2]])
x1 = np.random.multivariate_normal(mu_vec1, cov_mat1, 100)
mu_vec1 = mu_vec1.reshape(1,2).T # to 1-col vector

mu_vec2 = np.array([1,2])
cov_mat2 = np.array([[1,0],[0,1]])
x2 = np.random.multivariate_normal(mu_vec2, cov_mat2, 100)
mu_vec2 = mu_vec2.reshape(1,2).T

# 각 샘플에 Y 값 추가
y1 = np.zeros((100,1))
y2 = np.ones((100,1))
x1_y1 = np.concatenate((x1, y1), axis=1)
x2_y2 = np.concatenate((x2, y2), axis=1)

# 모든 데이터 결합
total = np.concatenate((x1_y1, x2_y2), axis=0)

plt.scatter(total[:,0], total[:,1], c=total[:,2])
plt.show()
```

```python
# pandas 이용해서 csv 파일 저장
pdx1 = pd.DataFrame(total)
pdx1.to_csv("random_sample_c_2.csv", header=False, index=False)
```

```python
# 왜곡 반영
total[:,0] = total[:,0] * 10000
#total[:,0] = total[:,0] + 2000

# pandas 이용해서 csv 파일 저장
pdx1 = pd.DataFrame(total)
pdx1.to_csv("random_sample_c_2_distortion.csv", header=False, index=False)

plt.scatter(total[:,0], total[:,1], c=total[:,2])
plt.show()
```
